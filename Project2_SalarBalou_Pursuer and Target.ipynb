```python
import numpy as np
from scipy.integrate import solve_ivp
import scipy.optimize as opt
from scipy.misc import derivative

import plotly.io as pio
from plotly.subplots import make_subplots
import plotly.graph_objects as go
pio.templates.default= 'plotly_white' 

def solve_plot(T, k, P0, tend):
  tn = np.linspace(0, tend, 1000)
  dt = tn[1] - tn[0]
  dTdt = derivative(T, dt)
  normTp = np.linalg.norm(dTdt)

  def rhs(t, P):
   dPdt = (k * normTp/np.linalg.norm(T(t) - P)) * (T(t) - P)
   return dPdt
 
  sol = solve_ivp(rhs, (0, tend), P0, method = 'Radau', dense_output=True)
  P = sol.sol
  solt, solP = sol.t, sol.y

  tplot = np.linspace(0, tend, 1000)

  distance = np.sqrt(np.sum((T(tplot) - P(tplot))**2, axis = 0))


  fig = make_subplots(rows=1, cols = 2)
  fig.add_trace(go.Scatter(x = T(tplot)[0], y = T(tplot)[1], name = 'Target'), row = 1, col = 1)
  fig.add_trace(go.Scatter(x = P(tplot)[0], y = P(tplot)[1], name = 'Pursuer'), row = 1, col = 1)
  fig.update_xaxes(nticks = 10, tickfont = dict(size = 18), ticks = 'outside', title = dict(text = 'X', font = dict(size = 20)), 
                  title_font = dict(size = 18), range = [-0.1, tplot[-1]], showgrid = False, linecolor = 'black', row =1, col =1)
  fig.update_yaxes(nticks = 10, tickfont = dict(size = 18), ticks = 'outside', title = dict(text = 'Y', font = dict(size = 20)), 
                  title_font = dict(size = 18), showgrid = False, linecolor = 'black', row =1, col =1)
 
  fig.add_trace(go.Scatter(x = tplot, y = distance, name = 'Distance'), row = 1, col = 2)
  fig.update_xaxes(nticks = 10, tickfont = dict(size = 18), ticks = 'outside', title = dict(text = 'time', font = dict(size = 20)), 
                  title_font = dict(size = 18), range = [-0.1, tplot[-1]], showgrid = False, linecolor = 'black', row =1, col =2)
  fig.update_yaxes(nticks = 10, tickfont = dict(size = 18), ticks = 'outside', title = dict(text = 'Distance', font = dict(size = 20)), 
                  title_font = dict(size = 18), showgrid = False, linecolor = 'black', row =1, col =2)
  fig.update_layout(height  = 600, width = 2000, legend = dict(yanchor ='top', y = 0.99, xanchor = 'left', x= 0.45))

  return fig
```

```python
def catch_plot(T, k, P0, tcatch):
  tn = np.linspace(0, tcatch, 1000)
  dt = tn[1] - tn[0]
  dTdt = derivative(T, dt)
  normTp = np.linalg.norm(dTdt)

  def rhs(t, P):
   dPdt = (k * normTp/np.linalg.norm(T(t) - P)) * (T(t) - P)
   return dPdt
 
  sol = solve_ivp(rhs, (0, tcatch), P0, method = 'Radau', dense_output=True)
  P = sol.sol
  solt, solP = sol.t, sol.y

  tplot = np.linspace(0, tcatch, 1000)
  distance = np.sqrt(np.sum((T(tcatch) - P(tcatch))**2, axis = 0))

  while (np.min(distance) > 1e-4):
   k += 1e-2
   sol = solve_ivp(rhs, (0, tcatch), P0, method = 'Radau', dense_output=True)
   P = sol.sol
   distance = np.sqrt(np.sum((T(tplot) - P(tplot))**2, axis = 0))

  distance = np.sqrt(np.sum((T(tplot) - P(tplot))**2, axis = 0))

  fig = make_subplots(rows=1, cols = 2)
  fig.add_trace(go.Scatter(x = T(tplot)[0], y = T(tplot)[1], name = 'Target'), row = 1, col = 1)
  fig.add_trace(go.Scatter(x = P(tplot)[0], y = P(tplot)[1], name = 'Pursuer'), row = 1, col = 1)
  fig.update_xaxes(nticks = 10, tickfont = dict(size = 18), ticks = 'outside', title = dict(text = 'X', font = dict(size = 20)), 
                  title_font = dict(size = 18), showgrid = False, linecolor = 'black', row =1, col =1)
  fig.update_yaxes(nticks = 10, tickfont = dict(size = 18), ticks = 'outside', title = dict(text = 'Y', font = dict(size = 20)), 
                  title_font = dict(size = 18), showgrid = False, linecolor = 'black', row =1, col =1)
 
  fig.add_trace(go.Scatter(x = tplot, y = distance, name = 'Distance'), row = 1, col = 2)
  fig.update_xaxes(nticks = 10, tickfont = dict(size = 18), ticks = 'outside', title = dict(text = 'time', font = dict(size = 20)), 
                  title_font = dict(size = 18), showgrid = False, linecolor = 'black', row =1, col =2)
  fig.update_yaxes(nticks = 10, tickfont = dict(size = 18), ticks = 'outside', title = dict(text = 'Distance', font = dict(size = 20)), 
                  title_font = dict(size = 18), showgrid = False, linecolor = 'black', row =1, col =2)
  fig.update_layout(height  = 600, width = 2000, legend = dict(yanchor ='top', y = 0.99, xanchor = 'left', x= 0.45), title =f'Min Distance = {np.min(distance)}, k = {k}')

  return fig
```
